```python
# ðŸ“š Titanic Dataset - EDA Template

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Data
df = pd.read_csv('/content/train.csv')  # Make sure train.csv is in the same folder OR give correct path

# Step 3: Basic Info
print("\n--- First 5 Rows ---")
print(df.head())

print("\n--- Info ---")
print(df.info())

print("\n--- Describe ---")
print(df.describe())

print("\n--- Missing Values ---")
print(df.isnull().sum())

# Step 4: Handling Missing Data
# Filling 'Age' missing values with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Filling 'Embarked' missing values with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Dropping 'Cabin' (too many missing)
df.drop(columns=['Cabin'], inplace=True)

# Step 5: Data Visualization

# 5.1 Univariate Analysis
print("\n--- Plotting Univariate (Histograms) ---")
df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()

# 5.2 Bivariate Analysis
print("\n--- Plotting Survival Count vs Sex ---")
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.show()

# 5.3 Pairplot
print("\n--- Pairplot ---")
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']])
plt.show()

# 5.4 Heatmap
print("\n--- Correlation Heatmap ---")
plt.figure(figsize=(10,8))
# Include only numerical features for correlation calculation
numerical_df = df.select_dtypes(include=np.number)
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.show()

# Step 6: Observations
# (WRITE YOUR OBSERVATIONS HERE)
# Example:
# - More females survived than males.
# - Pclass 1 passengers had higher survival rate.
# - Fare is positively correlated with Survival.

# Step 7: Conclusion
# (WRITE A SUMMARY OF WHAT YOU FOUND)

# Example Summary:
# "In the Titanic dataset, survival chances were higher for females, younger passengers,
# and those who paid higher fares. Class of travel (Pclass) also significantly impacted survival chances."

# ðŸŽ¯ Done!
```
